```python
!pip install firebase

from firebase import firebase
import json
from google.colab import drive
drive.mount('/content/drive')

# Firebase URL
firebase_url = 'https://hw2-tiger-default-rtdb.europe-west1.firebasedatabase.app/'

# Initialize the connection to Firebase
FBconn = firebase.FirebaseApplication(firebase_url, None)

# Read the JSON data from the file
json_file_path = '/content/drive/My Drive/Colab Notebooks/Homework2/Tiger.json'
with open(json_file_path, 'r') as file:
    data = json.load(file)

# Function to upload data to Firebase
def upload_to_firebase(data):
    for item in data:
        result = FBconn.post('/HW2-Tiger/', item)
        print(f'Data inserted with key: {result["name"]}')

# Call the function to upload data
upload_to_firebase(data)

```
